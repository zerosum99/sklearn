```python
import math as m

def f(x):
    return m.sin(x)

def g(x):
    return np.sin(x)

N, L = 100000, 2 * m.pi
x, dx = np.linspace(0, L, N), L / N
```

```python
diff = []
for i in x:
    diff.append((f(i + dx) - f(i)) / dx)
```

```python
diff
```

```python
diff_ = np.gradient(g(x), dx)
```

```python
diff_
```
